# Extract informations  from API 

```python
import wbgapi as wb
import pandas as pd
```

```python
wb.source.info()
```

```python
wb.series.info(q='growth')
```

```python

# Fetch GDP data from the World Bank API for the years 2018 to 2022 with labels
gdp = wb.data.DataFrame('NY.GDP.MKTP.CD', time=range(2018, 2023), labels=True)

# Sort the DataFrame by the "Country" column in alphabetical order and reset the index
gdp = gdp.sort_values(by='Country').reset_index()

# Calculate the mean GDP for each country across the years 2018 to 2022 and add it as a new column "GDB_MEAN"
gdp["GDP_MEAN"] = gdp[['YR2018', 'YR2019', 'YR2020', 'YR2021', 'YR2022']].mean(axis=1)

# Drop the column "economy" from the DataFrame
gdp = gdp.drop('economy', axis=1)

# Drop the columns for individual years (YR2018, YR2019, YR2020, YR2021, YR2022) from the DataFrame
gdp = gdp.drop(['YR2018', 'YR2019', 'YR2020', 'YR2021', 'YR2022'], axis=1)

# clean data: 
gdp.isna().sum()

gdp= gdp.dropna()


gdp
```

```python
# Save the DataFrame to a CSV file named "gdp_data.csv"
gdp.to_csv('gdp_data.csv', index=False)
```
